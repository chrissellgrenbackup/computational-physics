```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
R = 1.0
a_values = [0.5, 1.0, 2.0]  # example values of a
z_points = 500  # number of points in z

# Function to compute y(z; a, R)
def y_square(z, a, R):
    if a != 1:
        h = ((2*(3+a)/a)**(1/3)) * R
        l = 0.5 * h**(-a)
        y2 = (z / (2*l))**(2/(1+a)) - z**2
    else:
        h = 2*R
        l = 1/h
        y2 = z / l - z**2
    y2 = np.maximum(y2, 0)
    return np.sqrt(y2)

# Plot
plt.figure(figsize=(10,7))

for a in a_values:
    # Create z values from 0 to h(a)
    if a != 1:
        h = ((2*(3+a)/a)**(1/3)) * R
    else:
        h = 2*R
    z = np.linspace(0, h, z_points)
    y = y_square(z, a, R)
    plt.plot(z, y, label=f'a = {a}', linewidth=2)

# Make fonts larger for readability
plt.xlabel('z', fontsize=16)
plt.ylabel('y(z)', fontsize=16)
plt.title('Curves for different E field power laws', fontsize=18)
plt.legend(fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True)

plt.savefig("/Users/chris/Desktop/y_z_plot.png", dpi=300, bbox_inches='tight')
plt.show()
```
